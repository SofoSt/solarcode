# Solar Map Dataset

## Info

All the information comes from the Athens Geoportal that has a catalog of geospatial data of Municipality of Athenshttp://gis.cityofathens.gr/layers/athens_geonode_data:geonode:c40solarmap#

Σύμφωνα με το εγκεκριμένο το 2017 Σχέδιο Δράσης για το Κλίμα, στόχος είναι η προώθηση ηλιακών φωτοβολταϊκών (PV) σε ιδιωτικά κτίρια σε όλη την Αθήνα. Ο χάρτη απόδοσης ηλιακής ενέργειας για τις στέγες των κτηρίων της Αθήνας θα αποτελούσε χρήσιμο εργαλείο για τους πολίτες, ώστε να εξετάσουν την αποδοτικότητα που θα είχε η τοποθέτηση φωτοβολταϊκών συστημάτων στις στέγες των κτηρίων τους. Μετά από αίτημα του Δήμου Αθηναίων προς το Δίκτυο C40 (https://www.c40.org/), παραδόθηκε η μελέτη που αφορά στη δυνατότητα εγκατάστασης φωτοβολταϊκών συστημάτων στις στέγες των κτηρίων της Αθήνας. Στο πλαίσιο αυτής της μελέτης παρήχθησαν δεδομένα σχετικά με την απόδοση ηλιακής ενέργειας κάθε στέγης.

### Read

```python
url = 'https://drive.google.com/uc?id=1j2c094i1g65o2Ch9PcOtoiuLf4Xwr1ej'
output = 'c40solarmap.csv'
gdown.download(url, output, quiet=False)

# Read the CSV file into a DataFrame
df = pd.read_csv(output)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Renaming the columns
df.columns = [
    'Unique Feature Identifier', 'Secondary Feature Identifier', 'Roof Geometry',
    'Unique Geographic Identifier', 'Feature Classification Identifier', 'Building Identifier',
    'Height of Building in Meters', 'Type of Roof', 'Roof Area in Square Meters',
    'Type of Building', 'Feature Classification', 'Name of Building',
    'Total Solar Radiation per Unit Area', 'Direct Solar Radiation from Sun', 'Diffuse Solar Radiation from Sky',
    'Usable Area', 'Estimated Number of Solar Panels', 'Actual Area',
    'Approximate Act Area ', 'Power Output of PV System(KWP)',
    'Estimated Annual Energy Production', 'Estimated Annual CO2 Emission Reduction',
    'Efficiency of PV System(KWH_KWP)', 'Savings for Residential Electricity',
    'Savings for Non-Residential Electricity', 'Payback Period for Residential Installations',
    'Payback Period for Non-Residential Installations', 'Length of Roof Perimeter', 'Computed Area of Roof Shape'
]

# Verify the new column names
print(df.columns)
```

```python
# Replace null values in 'Type of Building' with 'Undefined'
df['Type of Building'] = df['Type of Building'].fillna('Undefined')
```

#### Extract building latitude and longtitude

```python
import geopandas as gpd
from shapely.wkt import loads

# EPSG:2100 for Greece
projection_epsg = 2100

wkt_polygon = building_info['Roof Geometry']

# Create a GeoDataFrame with a bounding box as its geometry, in the EPSG:2100 CRS
gdf = gpd.GeoDataFrame([1], geometry=[loads(wkt_polygon)], crs=f'EPSG:{projection_epsg}')

# Calculate the centroid of the bounding box (still in EPSG:2100)
centroid_projected = gdf.geometry.centroid

# Convert the centroid's coordinates to geographic CRS (EPSG:4326) to get latitude and longitude
centroid_geo = centroid_projected.to_crs(epsg=4326)

# Access the centroid's longitude and latitude
longitude = centroid_geo.x.iloc[0]
latitude = centroid_geo.y.iloc[0]

print(f"Latitude: {latitude}")
print(f"Longitude: {longitude}")
```

### Solar Panel Efficiency
Comparing the efficiency of solar panels across different types of buildings could highlight where solar installations are most effective. Efficiency can be gauged by Efficiency of PV System, which could be plotted against Type of Building.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Type of Building', y='Efficiency of PV System(KWH_KWP)', data=df)
plt.xticks(rotation=45)
plt.title('Solar Panel Efficiency by Building Type')
plt.xlabel('Building Type')
plt.ylabel('Efficiency of PV System(KWH_KWP)')
plt.tight_layout()
plt.show()
```

### CO2 Emission Reduction Potential
A key benefit of solar panels is their potential to reduce CO2 emissions. You could aggregate the Estimated Annual CO2 Emission Reduction by Type of Building to show which buildings offer the most significant reduction potential.

```python
co2_reduction_by_type = df.groupby('Type of Building')['Estimated Annual CO2 Emission Reduction'].sum().sort_values(ascending=False)
co2_reduction_by_type.plot(kind='bar', figsize=(12, 6), color='green')
plt.title('CO2 Emission Reduction Potential by Building Type')
plt.xlabel('Building Type')
plt.ylabel('CO2 Reduction (kg/year)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter out rows where 'Type of Building' is 'Undefined'
defined_buildings_df = df[df['Type of Building'] != 'Undefined']

# Group by 'Type of Building' and sum CO2 reduction, then sort in descending order
co2_reduction_by_type = defined_buildings_df.groupby('Type of Building')['Estimated Annual CO2 Emission Reduction'].sum().sort_values(ascending=False)

# Plotting
co2_reduction_by_type.plot(kind='bar', figsize=(12, 6), color='green')
plt.title('CO2 Emission Reduction Potential by Building Type (Excluding \'Undefined\')')
plt.xlabel('Building Type')
plt.ylabel('CO2 Reduction (kg/year)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Correlation Heatmap

Understanding the relationship between various numerical features can help identify which factors are most influential for solar panel efficiency and output.

```python
# Select numerical columns for the correlation matrix
numerical_columns = ['Height of Building in Meters', 'Roof Area in Square Meters', 'Total Solar Radiation per Unit Area',
                     'Usable Area', 'Estimated Number of Solar Panels', 'Power Output of PV System(KWP)',
                     'Estimated Annual Energy Production', 'Estimated Annual CO2 Emission Reduction', 'Efficiency of PV System(KWH_KWP)']

correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
